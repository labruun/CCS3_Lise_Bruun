```python
# attempt two at generating the statements

def generate_statements(list_of_lists):
    # Function to generate statements for each sublist
    def process_list(sublist):
        true_statements = []
        false_statements = []
        n = len(sublist)
        for i, current in enumerate(sublist):
            for j, other in enumerate(sublist):
                if i < j:
                    true_statements.append(f"The {current} was to the left of the {other}.")
                    false_statements.append(f"The {current} was to the right of the {other}.")
                elif i > j:
                    true_statements.append(f"The {current} was to the right of the {other}.")
                    false_statements.append(f"The {current} was to the left of the {other}.")
        return true_statements, false_statements

    # Process each list in the input and collect statements
    all_true_statements = []
    all_false_statements = []
    for sublist in list_of_lists:
        true_s, false_s = process_list(sublist)
        all_true_statements.extend(true_s) 
        all_false_statements.extend(false_s)

    return all_true_statements, all_false_statements

# Example usage
categories = ["humanoid_2", "things_2", "vehicles_2", "animals_2"]

cat = [humanoid_2, things_2, vehicles_2, animals_2]


for i in range(4):
    true_statements, false_statements = generate_statements(cat[i])
    
    print("# " + categories[i])
    print("true_" + categories[i] + " = " + str(true_statements))
    print("\n")
    
    print("false_" + categories[i] + " = " + str(false_statements))
    print("\n\n")
    
```
